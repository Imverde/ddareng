## parquet 파일로 저장하기

```python
import pandas as pd

df_2021 = pd.read_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2021.csv', encoding='cp949')
df_2022 = pd.read_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2022.csv', encoding='cp949')
df_2023 = pd.read_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2023.csv', encoding='cp949')
df_2024_1q = pd.read_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2024_1분기.csv', encoding='cp949')
df_2024_2q = pd.read_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2024_2분기.csv', encoding='cp949')
```

```python
df_2021.head()
```

```python
# '연도' 컬럼 만들에서 2021데이터 넣기
df_2021['연도'] = 2021

# '연도'을 첫 번째 위치로 이동
cols = df_2021.columns.tolist()
cols = ['연도'] + [col for col in cols if col != '연도']
df_2021 = df_2021[cols]

# '자치구'로 컬럼명 변경하기
df_2021 = df_2021.rename(columns={'구분':'자치구'})

df_2021
```

## 2022

```python
df_2022['연도'] = 2022

# '연도'을 첫 번째 위치로 이동
cols = df_2022.columns.tolist()
cols = ['연도'] + [col for col in cols if col != '연도']
df_2022 = df_2022[cols]

# '자치구'로 컬럼명 변경하기
df_2022 = df_2022.rename(columns={'구분':'자치구'})

df_2022
```

## 2023

```python
df_2023['연도'] = 2023

# '연도'을 첫 번째 위치로 이동
cols = df_2023.columns.tolist()
cols = ['연도'] + [col for col in cols if col != '연도']
df_2023 = df_2023[cols]

# '자치구'로 컬럼명 변경하기
df_2023 = df_2023.rename(columns={'구분':'자치구'})

df_2023
```

## 특정 열에 대해서만 평균내기

```python
# 평균을 내고자 하는 열 이름들
columns_to_average = ['고급휘발유', '보통휘발유', '자동차용경유', '실내등유']

# 두 데이터프레임에서 특정 열만 선택
df_2024_1q_selected = df_2024_1q[columns_to_average]
df_2024_2q_selected = df_2024_2q[columns_to_average]

# 선택된 열들의 평균 계산
df_2024 = (df_2024_1q_selected + df_2024_2q_selected) / 2
```

```python
df_2024
```

```python
df_2024['연도'] = 2024

# '연도'을 첫 번째 위치로 이동
cols = df_2024.columns.tolist()
cols = ['연도'] + [col for col in cols if col != '연도']
df_2024 = df_2024[cols]

df_2024
```

```python
df_2024.insert(1, '자치구', ['서울', '강남구', '강동구', '강북구', '강서구', 
                          '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', 
                          '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', 
                          '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'])
```

```python
df_2024
```

```python
df_2021 = df_2021.drop(index=[0])
df_2022 = df_2022.drop(index=[0])
df_2023 = df_2023.drop(index=[0])
df_2024 = df_2024.drop(index=[0])
```

```python
df_2024
```

## 2021~2024 데이터프레임 합치기

```python
df = pd.concat([df_2021, df_2022, df_2023, df_2024], ignore_index=True)
df
```

## 2021년~2024년2분기 합친 데이터프레임 csv파일로 저장하기

```python
df.to_csv('G:/내 드라이브/DataSet/서울시 자치구별 유가데이터(년도별)/주유소_지역별_평균판매가격_2021~2024_2분기.csv', encoding='utf-8-sig', index=False)
```
