```python
# 출근시간대 7시~9시
selected = df[(df['기준_시간대'] >= 700) & (df['기준_시간대'] <= 900)]
selected
```

```python
# 퇴근시간대 17시~19시
selected_1 = df[(df['기준_시간대'] >= 1700) & (df['기준_시간대'] <= 1900)]
selected_1
```

```python
# 출근 시간대 대여 데이터
start = selected[selected['집계_기준'] == '출발시간']
start
```

```python
# 퇴근 시간대 반납 데이터
end = selected_1[selected_1['집계_기준'] == '도착시간']
end
```

```python
# 출근 시간대 대여 횟수 카운트해서 'start' 데이터 프레임에 추가
rental_counts = start['시작_대여소_ID'].value_counts()
start.loc[:, '대여횟수'] = start['시작_대여소_ID'].map(rental_counts)
start
```

```python
# 퇴근 시간대 반납 횟수 카운트해서 'end' 데이터 프레임에 추가
return_counts= end['종료_대여소_ID'].value_counts()
end.loc[:, '반납횟수'] = end['종료_대여소_ID'].map(return_counts)
end
```

```python
# 출근 시간대 대여대여소 카운트한 데이터 프레임에서 '시작_대여소_ID', '대여횟수'만 뽑아서 데이터프레임에 저장
start_2 = start[['시작_대여소_ID', '대여횟수']]
start_2.reset_index(drop=True, inplace=True)
start_2
```

```python
# 퇴근 시간대 반납대여소 카운트한 데이터 프레임에서 '종료_대여소_ID', '반납횟수'만 뽑아서 데이터프레임에 저장
end_2 = end[['종료_대여소_ID', '반납횟수']]
end_2.reset_index(drop=True, inplace=True)
end_2
```
