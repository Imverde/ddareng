```python
import pandas as pd

station_info = em.Read_File(r'G:\내 드라이브\DataSet\_파킷 파일\서울시 공공자전거 대여소 정보\공공자전거 대여소 정보_통합본_최종.parquet')

year_from = 2021
year_to = 2024

# 빈 리스트를 준비하여 연도별 DataFrame을 추가
dataframes = []
dataframes2 = []

for year in range(year_from, year_to + 1):
    bike_rent = em.Read_File(r'G:\내 드라이브\DataSet\_파킷 파일\서울특별시 공공자전거 대여이력 정보\서울특별시 공공자전거 대여이력 정보_' + str(year) + '_preprocessed.parquet')
    display(bike_rent)
    # 그룹화 및 집계
    # grouped_df는 전체 시간, grouped_df2는 출퇴근 시간
    bike_rent_start = bike_rent[(bike_rent['대여시간'].astype(str).str[:2].isin(['07', '08'])) & (bike_rent['반납시간'].astype(str).str[:2].isin(['07', '08']))]
    display(bike_rent_start)
    grouped_df_start_rent = bike_rent_start.groupby(['대여대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '출근대여수'}).reset_index()
    grouped_df_start_rent = grouped_df_start_rent.rename(columns={'대여대여소번호': '대여소번호'})
    grouped_df_start_return = bike_rent_start.groupby(['반납대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '출근반납수'}).reset_index()
    grouped_df_start_return = grouped_df_start_return.rename(columns={'반납대여소번호': '대여소번호'})
    # display(grouped_df_start_rent)
    # display(grouped_df_start_rent.columns)
    # display(grouped_df_start_return)
    # display(grouped_df_start_return)
    grouped_df = pd.merge(grouped_df_start_rent, grouped_df_start_return, on='대여소번호', how='outer')

    bike_rent_end = bike_rent[((bike_rent['대여시간'].astype(str).str[:2].isin(['17', '18'])) & (bike_rent['반납시간'].astype(str).str[:2].isin(['17', '18'])))]

    grouped_df_end_rent = bike_rent_end.groupby(['대여대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '퇴근대여수'}).reset_index()
    grouped_df_end_rent = grouped_df_end_rent.rename(columns={'대여대여소번호': '대여소번호'})
    grouped_df_end_return = bike_rent_end.groupby(['반납대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '퇴근반납수'}).reset_index()
    grouped_df_end_return = grouped_df_end_return.rename(columns={'반납대여소번호': '대여소번호'})
    # display(grouped_df_end_rent)
    # display(grouped_df_end_return)
    grouped_df2 = pd.merge(grouped_df_end_rent, grouped_df_end_return, on='대여소번호', how='outer')

    # del bike_rent

    grouped_df = pd.merge(grouped_df, grouped_df2, on='대여소번호', how='outer')
    grouped_df = pd.merge(grouped_df, station_info[['대여소번호', '대여소명', '상세주소', '위도', '경도']], on='대여소번호', how='left')
    # grouped_df = grouped_df.rename(columns={'대여소명': '대여대여소명', '상세주소': '대여상세주소'})
    # grouped_df.drop('대여소번호', axis=1, inplace=True)

    # grouped_df2 = pd.merge(grouped_df2, station_info[['대여소번호', '대여소명', '상세주소']], on='대여소번호', how='left')
    # grouped_df2 = grouped_df2.rename(columns={'대여소명': '대여대여소명', '상세주소': '대여상세주소'})
    # # grouped_df2.drop('대여소번호', axis=1, inplace=True)

    pd.options.display.max_rows = 100

    display(grouped_df.head(100))
    # display(grouped_df2.head(100))

    grouped_df.to_parquet(r'G:\내 드라이브\DataSet\_파킷 파일\서울특별시 공공자전거 대여이력 정보\출퇴근대여반납차이_' + str(year) + '.parquet')
#     # grouped_df2.to_parquet(r'C:\Users\user\Desktop\나따릉이\대여이력\공공자전거 대여반납 그룹_' + str(year) + '_출퇴근.parquet')

# #     # 연도 칼럼 추가
    grouped_df['연도'] = year
# #     grouped_df2['연도'] = year
    
# #     # 리스트에 추가
    dataframes.append(grouped_df)
# #     dataframes2.append(grouped_df2)


# #     # # 결과 출력
# #     # # display(grouped_df[grouped_df['이용거리(M)'] > 0].sort_values(by='이용량', ascending=False).reset_index(drop=True))
# #     # display(grouped_df2.sort_values(by='이용거리(M)', ascending=False).reset_index(drop=True))
# #     # # display(grouped_df[grouped_df['이용거리(M)'] > 0].sort_values(by='이용시간(분)', ascending=False).reset_index(drop=True))

# #     # # display(grouped_df[grouped_df['대여대여소번호'] != grouped_df['반납대여소번호']].sort_values(by='이용량', ascending=False).reset_index(drop=True))
# #     # display(grouped_df2[grouped_df2['대여대여소번호'] != grouped_df2['반납대여소번호']].sort_values(by='이용거리(M)', ascending=False).reset_index(drop=True))
# #     # # display(grouped_df[grouped_df['대여대여소번호'] != grouped_df['반납대여소번호']].sort_values(by='이용시간(분)', ascending=False).reset_index(drop=True))

# #     # # display(grouped_df[grouped_df['이용거리(M)'] > 0].sort_values(by='이용량', ascending=False).reset_index(drop=True))
# #     # display(grouped_df2[grouped_df2['이용거리(M)'] > 0].sort_values(by='이용거리(M)', ascending=False).reset_index(drop=True))
# #     # # display(grouped_df[grouped_df['이용거리(M)'] > 0].sort_values(by='이용시간(분)', ascending=False).reset_index(drop=True))

# # # 리스트에 있는 모든 DataFrame을 하나로 합치기
combined_df = pd.concat(dataframes, ignore_index=True)
combined_df.to_parquet(r'G:\내 드라이브\DataSet\_파킷 파일\서울특별시 공공자전거 대여이력 정보\출퇴근대여반납차이_' + str(year_from) + '_' + str(year_to) + '.parquet')

# # combined_df2 = pd.concat(dataframes2, ignore_index=True)
# # combined_df2.to_parquet(r'C:\Users\user\Desktop\나따릉이\대여이력\공공자전거 대여반납 그룹_' + str(year_from) + '_' + str(year_to) + '_출퇴근.parquet')
```
