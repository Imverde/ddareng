## 4. NULL값 있는 '반납대여소번호' 컬럼, '이용거리(M)' 컬럼 확인하기

- '반납대여소번호' null값 확인

```python
df[df['반납대여소번호'].isnull()]
```

- '이용거리(M)' null값 확인

```python
df[df['이용거리(M)'].isnull()]
```

- '이용거리(M)' == 0.0인 개수 확인

```python
df[df['이용거리(M)'] == 0.0]
```

- '년월' 컬럼 생성 후, '반납대여소번호' 컬럼의 년월 분포 확인하기

```python
df['년월'] = df['대여일자'].str.slice(start = 0, stop = 7)
df.head()
```

```python
# df['대여일자'].str[:7].value_counts().sort_index()
```

```python
null_counts = df.loc[df['반납대여소번호'].isnull(), '년월'].value_counts()

import matplotlib.pyplot as plt

# 글씨체 깨지는 문제 해결

from matplotlib import font_manager, rc
# window에서 '맑은 고딕' 폰트를 사용하도록 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 그래프 그리기
plt.figure(figsize=(10,6))
null_counts.plot(kind='bar', color='skyblue')

# 그래프에 제목과 레이블 추가
plt.title('Null 반납대여소번호의 년월별 빈도수', fontsize=16)
plt.xlabel('년월', fontsize=12)
plt.ylabel('빈도수', fontsize=12)

# 그래프 출력
# x축 레이블을 45도 회전하여 보기 쉽게 설정
plt.xticks(rotation=45)
# x축 레이아웃을 조정하여 요소 간 간격 조정
plt.tight_layout()

# 그래프 위에 빈도수 표시
for i, value in enumerate(null_counts):
    plt.text(i, value, f'{value}', ha='center', va='bottom', fontsize=9)

plt.show()
```

- '반납대여소'가 Null 값이고 '이용거리(M)'가 0.0인 행 삭제하기

```python
df= df[~(df['반납대여소번호'].isnull() & (df['이용거리(M)'] == 0.0))]
```

```python
# 대여일자와 대여시간, 반납일자와 반납시간을 문자열로 변환하여 결합 후 datetiem 형식으로 변환
df['대여_시간'] = pd.to_datetime(df['대여일자'].astype(str) + ' ' + df['대여시간'].astype(str))
df['반납_시간'] = pd.to_datetime(df['반납일자'].astype(str) + ' ' + df['반납시간'].astype(str))

# 반납시간과 대여시간의 차이를 분 단위로 계산
df['이용_시간차이(분)'] = round((df['반납_시간'] - df['대여_시간']).dt.total_seconds() / 60, 2)
```
