## 1. 가설 1) 따릉이 최다 사용량이 출퇴근 시간이고 

### - 최다 사용량이 출퇴근 시간인 것을 확인한다.  

- 대여일자 컬럼을 datetime 형식으로 변환

```python
df_2024['대여일자'] = pd.to_datetime(df_2024['대여일자'], format='%Y-%m-%d')
# df_2020['대여시간'].astype('datetime64[ns]')
```

```python
print(df_2024['대여일자'].dtype)
```

```python
df_2024.head()
```

- datetime 형식으로 변경된 '대여일자' 컬럼에서 요일을 추출하여 새로운 컬럼으로 저장

```python
# dt.weekday : 요일 정보를 숫자로 추출 
# 0 : 월요일, 1 : 화요일, .... 6 : 일요일
df_2024['요일_인덱스'] = df_2024['대여일자'].dt.weekday
```

```python
df_2024.head()
```

```python
# 0~6번 인덱스 위치에 각 한글 요일 이름이 담긴 리스트
weekday_list = ['월', '화', '수', '목', '금', '토', '일']

# apply 함수로 숫자 -> 요일 이름으로 변환
df_2024['요일'] = df_2024.apply(lambda x : weekday_list[x['요일_인덱스']], axis=1)
```

## 출근 시간대 top20 대여소 지도에 나타내기 

```python
import pandas as pd
import numpy as np
import json

import folium
```

```python
# 나의 루트 경로 설정
path_dateset = 'G:/내 드라이브/DataSet/'
```

```python
cols = ['대여소번호', '보관소(대여소)명', '자치구', '상세주소', '위도', '경도', '설치시기', '거치대수(LCD)', '거치대수(QR)', '운영방식']
rental = pd.read_excel(path_dateset + '_최종 병합 파일\서울시 공공자전거 대여소 정보/공공자전거 대여소 정보(24.6월 기준).xlsx', sheet_name='대여소현황', names = cols)
rental = rental.iloc[4:]
rental.reset_index(drop = True, inplace = True)
```

```python
rental.head(3)
```

```python
# top_20_commute_data 정렬
top_20_map = top_20_commute_data.sort_values(by='이용건수', ascending=False)

# top 20 대여소명 목록 생성
top_20_stations = top_20_map['대여소명_y'].tolist()

# rental DataFrame에서 top 20 대여소 정보 추출
filtered_rental = rental[rental['보관소(대여소)명'].isin(top_20_stations)]

# 지도 생성
center = [37.541, 126.986]
bike_map = folium.Map(location=center, zoom_start=13)

# 최대 이용건수 계산 (마커 크기 조절에 사용)
max_usage = top_20_map['이용건수'].max()

# 자전거 대여소 표시
for i, row in filtered_rental.iterrows():
    # top_20_commute_data에서 해당 대여소의 이용건수 찾기
    usage = top_20_map[top_20_map['대여소명_y'] == row['보관소(대여소)명']]['이용건수'].values[0]
    
    # 이용건수에 따라 마커 크기 조절 (최소 5, 최대 15)
    marker_size = 5 + (usage / max_usage) * 10
    
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=marker_size,
        popup=f"대여소명_y: {row['보관소(대여소)명']}<br>이용건수: {usage}",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(bike_map)

# 지도 저장
bike_map.save("top_20_bike_stations.html")
```
