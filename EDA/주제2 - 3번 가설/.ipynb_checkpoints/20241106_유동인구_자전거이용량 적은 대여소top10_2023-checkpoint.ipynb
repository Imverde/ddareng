```python
import time
import pandas as  pd

filepath = 'G:\\내 드라이브\\DataSet\\_파킷 파일\\서울특별시 공공자전거 대여이력 정보\\'
filename = '서울특별시 공공자전거 대여이력 정보_' + _year + '_preprocessed.parquet'
start_time = time.time()
df = pd.read_parquet(filepath + filename)
print("read_parquet time :", time.time() - start_time)
df.head(5)
```

### 반납대여소 null값 삭제안함.

```python
df[df.isnull().any(axis=1)]
```

```python
df_rent = df.groupby(['대여대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '대여수'}).reset_index()
df_rent = df_rent.rename(columns={'대여대여소번호': '대여소번호'})
df_rent
```

```python
df_return = df.groupby(['반납대여소번호']).agg({
        '자전거번호': 'count'  # count를 이용해 이용량 계산
    }).rename(columns={'자전거번호': '반납수'}).reset_index()
df_return = df_return.rename(columns={'반납대여소번호': '대여소번호'})
df_return
```

```python
use = pd.merge(df_rent, df_return, on='대여소번호', how='left')
use
```

```python
use['총이용량'] = use['대여수'] + use['반납수']
use
```

```python
use.loc[use['대여소번호'] == '01868']
```

```python
lowest_usage_12 = use.nsmallest(12, '총이용량')
lowest_usage_12
```

```python
merge = pd.merge(lowest_usage_12, rental, on='대여소번호', how='left')
merge
```

### 첫번째(01816), 두번째(04314)로 이용량 적은 대여소. 대여소 정보 없음. 2024년 6월 대여소 정보에 없음.
### 따라서 사라진 것으로 추정되어 제외함

```python
merge_2 = merge.drop(index=[0,1])
merge_2.reset_index()
merge_2
```

```python
merge_2['합친컬럼'] = merge_2['대여소명'] + '\n' +' (' + merge_2['자치구'] + ') '
merge_2
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 한글 폰트 설정 (Windows: Malgun Gothic)
plt.rcParams['font.family'] = 'Malgun Gothic'
# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 막대 그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(merge_2['합친컬럼'], merge_2['총이용량'], color='skyblue')
plt.xlabel('대여소명')
plt.ylabel('총이용량')
plt.title('2023년 대여소별 총이용량 적은 10개 대여소')
plt.xticks(rotation=45, ha='right')

# Add 'Total Usage' labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height,
             f'{height:,.0f}',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
